# Try pydicom

```python
import pydicom

ds = pydicom.dcmread(path_rtdose, stop_before_pixels=True)  # faster: skip pixel data

# 1) Access by keyword (attribute-style)
print(ds.StudyDate)          # value
print(ds.SOPClassUID)        # value

# 2) Access the DataElement by keyword, then read its metadata
el = ds.data_element("SOPClassUID")
print(el.keyword)            # 'SOPClassUID'
print(el.tag)                # (0008,0016) as a Tag object
print(el.tag.group, el.tag.element)  # 8 22 (seem to be hexadecimal/base-16)
print(hex(el.tag.group), hex(el.tag.element))  # 0x8 0x16， 0x is prefix for 16 hexadecimal
print(el.VR, el.VM)          # e.g., 'UI', 1

# 3) Access by tag directly
el2 = ds[(0x0008, 0x0016)]
print(el2.keyword, el2.value)
```
