```python
for ci, contour in enumerate(contour_data.ContourSequence):
    contour_points = np.array(contour.ContourData).reshape(-1, 3)  # (x, y, z) in mm
    print(f"Contour {ci}: {contour_points.shape[0]} points")
    print(f"Contour points shape: {contour_points.shape}")
    print(f"\nContour points:\n{contour_points}")

    # Compute distances between adjacent points
    num_points = contour_points.shape[0]
    distances = np.zeros(num_points)
    
    # Compute distances: 1-2, 2-3, ..., (n-1)-n
    for i in range(num_points - 1):
        distances[i] = np.linalg.norm(contour_points[i+1] - contour_points[i])
    
    # Compute distance from last point to first point (wrapping): n-1
    distances[num_points - 1] = np.linalg.norm(contour_points[0] - contour_points[num_points - 1])
    
    print(f"\nDistances between adjacent points (in mm):")
    print(f"Total number of distances: {len(distances)}")
    for i in range(num_points):
        if i < num_points - 1:
            print(f"  Point {i+1} to {i+2}: {distances[i]:.4f} mm")
        else:
            print(f"  Point {i+1} to 1 (wrapping): {distances[i]:.4f} mm")
    
    print(f"\nDistance statistics:")
    print(f"  Min distance: {np.min(distances):.4f} mm")
    print(f"  Max distance: {np.max(distances):.4f} mm")
    print(f"  Mean distance: {np.mean(distances):.4f} mm")
    print(f"  Std distance: {np.std(distances):.4f} mm")
    print(f"\nDistance array: {distances}")

    break
```

```python

from skimage.draw import polygon2mask

verbose = True


slc_cnt = 0

# --- 1) Convert each contour to index space and group by slice --- #
for ci, contour in enumerate(contour_data.ContourSequence):
    contour_points = np.array(contour.ContourData).reshape(-1, 3)  # (x, y, z) in mm

    # Transform physical coordinates to voxel indices for this image
    indices = np.array(
        [image_reader.image.TransformPhysicalPointToIndex(tuple(pt)) for pt in contour_points]
    )  # (N, 3) in (i, j, k) ~ (x, y, z indices)

    # Swap x & y so indices[:,0] = row (y), indices[:,1] = col (x)
    indices[:, [1, 0]] = indices[:, [0, 1]]

    # All points for this contour should lie on the same slice (same k index)
    slice_index = np.unique(indices[:, 2])
    if len(slice_index) != 1:
        raise ValueError("Contour points do not lie on a single slice")
    slice_index = int(slice_index[0])


    # Convert indices to 2D for the slice
    points_2d = indices[:, :2]

    # Create a boolean mask for the contour
    slice_shape = mask[slice_index].shape
    slice_mask = polygon2mask(slice_shape, points_2d)

    print(f"ci: {ci}, slice_index: {slice_index}, np.sum(slice_mask): {np.sum(slice_mask)}")

    slc_cnt += np.sum(slice_mask)


    # plt.figure()
    # plt.imshow(slice_mask)
    # plt.title(str(slice_index))

    # Update the mask array with this slice mask
    mask[slice_index] = np.maximum(mask[slice_index], slice_mask.astype(np.uint8))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import SimpleITK as sitk
from skimage.draw import polygon2mask

# ==========================================
# 1. Geometry Setup
# ==========================================

target_spacing = (0.5, 0.5, 0.5)

# Get Original Image properties
orig_image = image_reader.image
orig_size = np.array(orig_image.GetSize())       # (x, y, z)
orig_spacing = np.array(orig_image.GetSpacing()) # (x, y, z)
orig_origin = orig_image.GetOrigin()
orig_direction = orig_image.GetDirection()

print(f"Original Geometry: Size{orig_size} | Spacing{orig_spacing}")

# --- STEP A: Create Intermediate "Anisotropic" Grid ---
# High res in X/Y (0.5mm), but Original res in Z (e.g. 3mm)
# This ensures every Z-slice in the original image maps to a Z-slice here.
inter_spacing = (target_spacing[0], target_spacing[1], orig_spacing[2])
inter_size = np.round(orig_size * orig_spacing / inter_spacing).astype(int)

# Ensure Z size matches original exactly to avoid rounding errors dropping the last slice
inter_size[2] = orig_size[2]

ref_img_inter = sitk.Image(inter_size.tolist(), sitk.sitkUInt8)
ref_img_inter.SetOrigin(orig_origin)
ref_img_inter.SetSpacing(inter_spacing)
ref_img_inter.SetDirection(orig_direction)

# --- STEP B: Create Final "Isotropic" Grid ---
# This is the destination 0.5mm isotropic grid
final_size = np.round(orig_size * orig_spacing / target_spacing).astype(int)
ref_img_final = sitk.Image(final_size.tolist(), sitk.sitkUInt8)
ref_img_final.SetOrigin(orig_origin)
ref_img_final.SetSpacing(target_spacing)
ref_img_final.SetDirection(orig_direction)

# ==========================================
# 2. Rasterize to Intermediate (Anisotropic) Volume
# ==========================================

# Numpy array for intermediate volume (z, y, x)
mask_inter_arr = np.zeros(inter_size[::-1], dtype=np.uint8)

print("Rasterizing contours to intermediate stack...")

for contour in contour_data.ContourSequence:
    contour_points_phys = np.array(contour.ContourData).reshape(-1, 3)
    
    # Transform to Intermediate Continuous Indices
    indices_inter = np.array(
        [ref_img_inter.TransformPhysicalPointToContinuousIndex(tuple(pt)) for pt in contour_points_phys]
    )
    
    # Identify Z-slice (should match original image slice indices now)
    z_indices = indices_inter[:, 2]
    slice_index_z = int(np.round(np.median(z_indices)))
    
    if 0 <= slice_index_z < mask_inter_arr.shape[0]:
        # Rasterize 2D polygon on this slice
        points_2d_yx = indices_inter[:, [1, 0]]
        slice_shape_yx = (mask_inter_arr.shape[1], mask_inter_arr.shape[2])
        slice_mask = polygon2mask(slice_shape_yx, points_2d_yx)
        
        # Accumulate (Logical OR)
        mask_inter_arr[slice_index_z] = np.maximum(mask_inter_arr[slice_index_z], slice_mask.astype(np.uint8))

# Convert numpy array back to SITK Image for resampling
mask_inter_img = sitk.GetImageFromArray(mask_inter_arr)
mask_inter_img.CopyInformation(ref_img_inter)

# ==========================================
# 3. Resample to High-Res Isotropic Volume (Interpolation)
# ==========================================

print("Resampling (interpolating) to 0.5mm isotropic grid...")

# We use Linear interpolation. This creates gray values (0.0 to 1.0) between binary slices.
# This effectively "lofts" the shape between the separated slices.
resampler = sitk.ResampleImageFilter()
resampler.SetReferenceImage(ref_img_final)
resampler.SetInterpolator(sitk.sitkLinear) 
resampler.SetDefaultPixelValue(0)
resampler.SetTransform(sitk.Transform()) # Identity transform

# Execute Resampling
mask_hr_gray = resampler.Execute(mask_inter_img)

# Threshold back to Binary (> 0.5)
mask_hr_binary = sitk.BinaryThreshold(mask_hr_gray, lowerThreshold=0.5, upperThreshold=1.1, insideValue=1, outsideValue=0)

# Convert to numpy for visualization (z, y, x)
mask_hr = sitk.GetArrayFromImage(mask_hr_binary)

print(f"Final Isotropic Mask Shape: {mask_hr.shape}")

# ==========================================
# 4. Visualization (Updated Logic)
# ==========================================
# Note: The visualization loop logic remains largely the same, 
# but now 'mask_hr' contains interpolated data between the original slices.

# --- Simple Helper for FOV calc ---
fov_x_mm = orig_size[0] * orig_spacing[0]
fov_y_mm = orig_size[1] * orig_spacing[1]
fov_z_mm = orig_size[2] * orig_spacing[2]

def plot_orthogonal_view(slice_idx, orientation, mr_slice, def_mask_slice, hr_mask_slice, extent):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(f"{orientation} View")

    # Common settings
    for ax in axs:
        ax.imshow(mr_slice, cmap='gray', aspect='auto', extent=extent, origin='upper')

    # 1. MR Only
    axs[0].set_title("MR Image")

    # 2. MR + Default Mask
    if np.any(def_mask_slice):
        axs[1].contour(def_mask_slice, levels=[0.5], colors='red', linewidths=1, extent=extent, origin='upper')
    axs[1].set_title("Default Resolution Contour")

    # 3. MR + High-Res Mask
    if np.any(hr_mask_slice):
        # Use 'lime' to distinguish. This should now appear as solid lines in Sag/Cor views!
        axs[2].contour(hr_mask_slice, levels=[0.5], colors='lime', linewidths=0.5, extent=extent, origin='upper')
    axs[2].set_title("High Res (Interpolated) Contour")
    
    plt.show()

# --- Sagittal Example (Middle of structure) ---
# Calculate the High-Res Image object for coordinate math
img_hr_obj = mask_hr_binary

print("Visualizing Sagittal View...")
sag_idx = np.argmax(np.sum(mask_default, axis=(0, 1))) # Find slice with max content
mr_sag = sitk.GetArrayFromImage(orig_image)[:, :, sag_idx]
def_sag = mask_default[:, :, sag_idx]

# Find corresponding slice in HR
phys_x = orig_image.TransformIndexToPhysicalPoint((int(sag_idx), 0, 0))[0]
hr_x_idx = int(np.round(img_hr_obj.TransformPhysicalPointToContinuousIndex((phys_x, 0, 0))[0]))
hr_sag = mask_hr[:, :, hr_x_idx] if 0 <= hr_x_idx < mask_hr.shape[2] else np.zeros_like(mr_sag)

plot_orthogonal_view(sag_idx, "Sagittal", mr_sag, def_sag, hr_sag, [0, fov_y_mm, fov_z_mm, 0])

# --- Coronal Example ---
print("Visualizing Coronal View...")
cor_idx = np.argmax(np.sum(mask_default, axis=(0, 2))) 
mr_cor = sitk.GetArrayFromImage(orig_image)[:, cor_idx, :]
def_cor = mask_default[:, cor_idx, :]

# Find corresponding slice in HR
phys_y = orig_image.TransformIndexToPhysicalPoint((0, int(cor_idx), 0))[1]
hr_y_idx = int(np.round(img_hr_obj.TransformPhysicalPointToContinuousIndex((0, phys_y, 0))[1]))
hr_cor = mask_hr[:, hr_y_idx, :] if 0 <= hr_y_idx < mask_hr.shape[1] else np.zeros_like(mr_cor)

plot_orthogonal_view(cor_idx, "Coronal", mr_cor, def_cor, hr_cor, [0, fov_x_mm, fov_z_mm, 0])
```
