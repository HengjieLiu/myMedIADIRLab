```python
# ============================================================================
# LOCAL DROP-IN FUNCTIONS: Convert RTStructure contour_data to mask
# ============================================================================
# This cell contains local implementations of the mask conversion logic
# following the slice-by-slice polygon2mask approach
# ============================================================================

import numpy as np
import SimpleITK as sitk
from skimage.draw import polygon2mask
from pathlib import Path

def process_single_contour_to_mask(contour, sitk_image, mask_array):
    """
    Process a single contour and update the mask array slice-by-slice.
    
    This is the core function that converts one contour polygon to a 2D mask
    on its corresponding slice, then combines it with existing masks on that slice.
    
    The key idea: Each contour belongs to a single slice (z-coordinate).
    We convert the 3D contour points to pixel space, extract the 2D polygon,
    use polygon2mask to fill the polygon, and combine with existing masks.
    
    Args:
        contour: pydicom contour dataset containing ContourData
                ContourData is a flat list: [x1, y1, z1, x2, y2, z2, ...]
        sitk_image: SimpleITK.Image - The CT/MR image for coordinate transformation
        mask_array: numpy.ndarray - The 3D mask array to update (shape: [z, y, x])
    
    Returns:
        None - The mask_array is modified in-place
    
    Details:
        1. Extract contour points: ContourData is reshaped into (N, 3) array
        2. Transform physical to pixel space: Use SimpleITK's TransformPhysicalPointToIndex
        3. Swap x/y: Convert from (x, y, z) to (col, row, slice) indexing
        4. Validate single slice: Ensure all points are on the same z-slice
        5. Extract 2D polygon: Get (col, row) coordinates for the slice
        6. Create polygon mask: Use polygon2mask to fill inside the polygon
        7. Combine masks: Use np.maximum to union with existing mask (logical OR)
    """
    # Step 1: Extract contour points from DICOM ContourData
    # ContourData is a flat list: [x1, y1, z1, x2, y2, z2, ...]
    # Reshape to (N, 3) array where each row is (x, y, z) in physical space (mm)
    contour_points_physical = np.array(contour.ContourData).reshape(-1, 3)
    
    # Step 2: Transform contour points from physical space (mm) to pixel space (indices)
    # SimpleITK's TransformPhysicalPointToIndex converts physical coordinates
    # to pixel indices. This accounts for image spacing, origin, and orientation.
    pixel_indices = []
    for pt in contour_points_physical:
        # Transform (x, y, z) in physical space to (x_idx, y_idx, z_idx) in pixel space
        pixel_index = sitk_image.TransformPhysicalPointToIndex(tuple(pt))
        pixel_indices.append(pixel_index)
    
    pixel_indices = np.array(pixel_indices)  # Shape: (N, 3) where columns are [x, y, z] indices
    
    # Step 3: Swap x and y to match numpy array indexing convention
    # SimpleITK uses (x, y, z) but numpy arrays use (row, col) = (y, x)
    # So we swap: (x_idx, y_idx, z_idx) -> (y_idx, x_idx, z_idx) = (row, col, slice)
    pixel_indices[:, [1, 0]] = pixel_indices[:, [0, 1]]
    # Now: pixel_indices[:, 0] = row indices (y)
    #      pixel_indices[:, 1] = col indices (x)  
    #      pixel_indices[:, 2] = slice indices (z)
    
    # Step 4: Validate that all contour points lie on a single slice
    # Extract unique slice indices (z-coordinates)
    slice_indices = np.unique(pixel_indices[:, 2])
    if len(slice_indices) != 1:
        raise ValueError(
            f"Contour points do not lie on a single slice. "
            f"Found slices: {slice_indices}"
        )
    
    # Get the slice index (z-coordinate) for this contour
    slice_idx = int(slice_indices[0])
    
    # Step 5: Extract 2D polygon coordinates for this slice
    # We only need (row, col) coordinates since all points are on the same slice
    polygon_2d = pixel_indices[:, :2]  # Shape: (N, 2) with columns [row, col]
    
    # Step 6: Create a 2D boolean mask by filling the polygon
    # polygon2mask fills the interior of the polygon defined by polygon_2d
    # on a canvas of size slice_shape
    slice_shape = mask_array[slice_idx].shape  # Shape: (height, width) = (rows, cols)
    slice_mask_2d = polygon2mask(slice_shape, polygon_2d)  # Returns boolean array
    
    # Step 7: Combine this contour's mask with existing masks on the same slice
    # np.maximum performs element-wise maximum (logical OR for binary masks)
    # If either mask has 1 at a pixel, the result is 1
    # This handles multiple contours on the same slice (e.g., separate regions)
    mask_array[slice_idx] = np.maximum(
        mask_array[slice_idx],           # Existing mask on this slice
        slice_mask_2d.astype(np.uint8)   # New mask from this contour (convert bool to uint8)
    )


def create_mask_from_contour_data(contour_data, sitk_image):
    """
    Create a 3D binary mask from RTStructure contour data.
    
    This is the main function that orchestrates the slice-by-slice mask creation.
    It processes all contours in the contour_data and combines them into a single 3D mask.
    
    The key idea: Process each contour independently, where each contour belongs to
    one slice. Use polygon2mask to fill each contour on its slice, then combine
    all contours to build the complete 3D mask.
    
    Args:
        contour_data: pydicom Sequence containing ContourSequence
                     Each item in ContourSequence is a contour dataset
        sitk_image: SimpleITK.Image - The CT/MR image for coordinate transformation
    
    Returns:
        numpy.ndarray: 3D binary mask array with shape (z, y, x) = (slices, rows, cols)
                      Values: 1 = inside structure, 0 = outside structure
    
    Details:
        1. Initialize 3D mask: Create zero array matching CT image shape
        2. Process each contour: Loop through all contours in ContourSequence
        3. Slice-by-slice filling: Each contour fills its slice using polygon2mask
        4. Combine contours: Multiple contours on same slice are combined with np.maximum
        5. Return final mask: Complete 3D mask representing the structure
    """
    # Step 1: Get the shape of the CT/MR image to create matching mask
    # SimpleITK images have shape (width, height, depth) in (x, y, z)
    # But numpy arrays use (depth, height, width) = (z, y, x)
    image_array = sitk.GetArrayFromImage(sitk_image)
    mask_shape = image_array.shape  # Shape: (z, y, x) = (slices, rows, cols)
    
    # Initialize 3D mask with zeros (all pixels initially outside structure)
    mask_3d = np.zeros(mask_shape, dtype=np.uint8)
    
    # Step 2: Process each contour in the ContourSequence
    # Each contour is a polygon defined on a single slice
    # Multiple contours may exist on the same slice (e.g., separate regions)
    num_contours = len(contour_data.ContourSequence)
    print(f"Processing {num_contours} contours to create mask...")
    
    for i, contour in enumerate(contour_data.ContourSequence):
        try:
            # Process this contour and update the mask array in-place
            process_single_contour_to_mask(contour, sitk_image, mask_3d)
        except Exception as e:
            print(f"Warning: Failed to process contour {i+1}/{num_contours}: {e}")
            continue
    
    # Step 3: Return the complete 3D mask
    # The mask now contains 1s where the structure exists, 0s elsewhere
    num_voxels = np.sum(mask_3d > 0)
    print(f"Mask created: {num_voxels} voxels marked as structure")
    
    return mask_3d


# ============================================================================
# Load the referenced CT/MR image for coordinate transformation
# ============================================================================
# We need the image to transform physical coordinates to pixel coordinates
# Find the referenced image series from the RTStructure
# ============================================================================

# Find the referenced image series from RTStructure
# RTStructure references the image series it was drawn on
rtstruct_dataset = rt_reader.rtstruct

# Get the referenced frame of reference and series information
referenced_frame_of_ref = rtstruct_dataset.ReferencedFrameOfReferenceSequence[0]
referenced_study = referenced_frame_of_ref.RTReferencedStudySequence[0]
referenced_series = referenced_study.RTReferencedSeriesSequence[0]
series_instance_uid = referenced_series.SeriesInstanceUID

print(f"Referenced Series Instance UID: {series_instance_uid}")

# Find the image directory (should be in the same study directory)
rt_struct_path_obj = Path(rt_struct_path)
study_dir = rt_struct_path_obj.parent.parent  # Go up to study directory

print(f"Searching for image series in: {study_dir}")

# Look for the image series directory
image_reader = None
for subdir in study_dir.iterdir():
    if subdir.is_dir():
        try:
            # Try MR modality first (since these are MR studies)
            image_reader = DICOMImageReader(
                str(subdir),
                modality="MR",
                series_instance_uid=series_instance_uid
            )
            image_reader.read()
            print(f"Found MR image series in: {subdir}")
            break
        except Exception:
            # Try CT modality if MR fails
            try:
                image_reader = DICOMImageReader(
                    str(subdir),
                    modality="CT",
                    series_instance_uid=series_instance_uid
                )
                image_reader.read()
                print(f"Found CT image series in: {subdir}")
                break
            except Exception:
                continue

if image_reader is None:
    raise ValueError("Could not find referenced image series. Cannot create mask without image.")

# ============================================================================
# Convert target1 contour_data to mask using local functions
# ============================================================================

# Get the contour_data for target1 (already extracted in previous cells)
# contour_data is a pydicom Sequence containing ContourSequence
# Each item in ContourSequence is a contour (polygon) on a single slice

# Convert contour_data to 3D mask using our local functions
target1_mask = create_mask_from_contour_data(contour_data, image_reader.image)

print(f"\nTarget1 mask created successfully!")
print(f"Mask shape: {target1_mask.shape} (z, y, x)")
print(f"Mask dtype: {target1_mask.dtype}")
print(f"Mask value range: [{target1_mask.min()}, {target1_mask.max()}]")
print(f"Number of slices with structure: {np.sum(np.any(target1_mask > 0, axis=(1, 2)))}")
print(f"Total voxels in structure: {np.sum(target1_mask > 0)}")
```
